# Netflix Recomendation Module:

```python
import pandas as pd    #Importing the pandas module as pd
```

```python
Movies = pd.read_csv('movies.csv')    #using the pd.read_csv reading the movies.csv file and storing into the Movies variable
Movies.head(10)
```

```python
Movies.columns  #columns names of a Movies Data Frame
```

```python
#Creating a Content column by concatinating the title,overview,keywords,cast,director columns
Movies['content'] = Movies['title']+''+ Movies['overview']+''+ Movies['keywords']+''+ Movies['cast']+''+ Movies['director']
```

```python
Movies['content']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer   #Importing the TfidVectorizer from sklearn
vectorizer = TfidfVectorizer(max_features=1000)
movie_vectors = vectorizer.fit_transform(Movies['content'].values.astype('U'))
```

```python
movie_vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity  #Importing the cosine_similarity from sklearn
similarity = cosine_similarity(movie_vectors) #Finding the cosine_similarity in movie_vectors and store into similarity variable
```

```python
similarity
```

```python
#Creating the recomendation function 
def recomendation(Movies):
    #Find Movie index with dataset
    Movies_index = Movies[Movies['title']==Movies].index[3]#The index[3] defines the 4th movie name in the dataset
    
    #Finding cosine similarity of movies
    distance = similarity[Movies_index]
    
    #sorting cosine similarity
    Movies_list =sorted(list(enumerate(distance)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in Movies_list:
        print(Movies.iloc[i[0]].title)
    
    
    
```

```python
recomendation(Movies) 
```
